```python
import pandas as pd
```

```python
onsite_go=pd.read_excel(r'C:/Users/Dell/Documents/Onsite_go.xlsx')
```

```python
onsite_go.dtypes
```

```python
onsite_go['total_sale']=onsite_go['TV']+onsite_go['Ref']+onsite_go['AC']+onsite_go['WP']
```

```python
onsite_go
```

```python
#month column creation
onsite_go['Month']=onsite_go['Sales date'].astype(str).apply(lambda x: x[:7])
```

```python
#cumsum column creation based group by month
onsite_go['cum_sum']=onsite_go.groupby(['Month'])['total_sale'].cumsum()
```

```python
# which date sales becomes greater than 5000
df=onsite_go.groupby(['Month']).apply(lambda x: x[x['cum_sum'] >5000][['Sales date','cum_sum']]).reset_index()
df.groupby('Month')[['Sales date','cum_sum']].first().reset_index()
```

```python
# top 5 sales date 
onsite_go.sort_values(by='total_sale',ascending=False)[['Sales date','total_sale']][:5]
```

```python
# bottom 5 sales date 
onsite_go.sort_values(by='total_sale',ascending=True)[['Sales date','total_sale']][:5]
```

```python
max_dt=int(str(max(onsite_go['Sales date']))[8:10])
```

```python
from datetime import datetime as dt
onsite_go['Sales date'].dt.day
```

```python
#compare two months sales
df1=onsite_go.groupby(['Month']).apply(lambda x: x[x['Sales date'].dt.day<=max_dt][['Sales date','cum_sum']])
df1.groupby('Month')[['Sales date','cum_sum']].last().reset_index()
```
