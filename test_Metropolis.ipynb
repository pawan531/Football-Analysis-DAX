```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/sakshambhatia/python_sample/main/python_sample.csv')
```

```python
df.head()
```

```python
'''def conct_str(str1):
    return 'vm'+str1

def conct_str1(a,b,c):
    return 'vm'+str(a)+str(b)+c
'''

df['Family_id']=df['MobileNumber'].astype(str).apply(lambda x:'+91'+x)
col=['PatientId','Age','Gender','MobileNumber']
df['Individual_id']=df[col].apply(lambda x:''.join(x.values.astype(str)),axis=1)
df['VisitDate']=pd.to_datetime(df['VisitDate'])






```

```python
df.head()
```

```python
# no of test done for familyId
df.groupby(['Family_id'])['Family_id'].count().sort_values(ascending=False)
```

```python
# Sum of revenue Generated for FamilyID
df.groupby('Family_id')['NetRevenue'].sum().sort_values(ascending=False).reset_index()
```

```python
# Frequency for Visits for Family ID
df.groupby('Family_id')['VisitNumber'].nunique().sort_values(ascending=False).reset_index()
```

```python
# first visits for Family ID
df.groupby('Family_id')['VisitDate'].first().reset_index()
```
