```python
def In_bucket(x):
    if x>1000000:
        return '>1000000'
    if x>500000 and x<1000000:
        return '500000-1000000'
    if x>200000 and x<500000:
        return '200000-500000'
    if x>100000 and x<200000:
        return '100000-200000'
    else:
        return '<100000'

    
cust_data['Income_bucket']=cust_data['MonthlyIncome'].apply(In_bucket)
    
    
    
```

```python
cust_data[cust_data['MonthlyIncome']>1000000].head()
```

```python
def In_age(x):
    if x<=22:
        return 'Student'
    if x>=23 and x<27:
        return 'UnMarried_Employed'
    if x>=27 and x<=40:
        return 'Married_with_Young_Kids'
    if x>40 and x<=60:
        return 'Pre_Retired'
    if x>60:
        return 'Retired'

cust_demo['Life_Stage_bucket']=cust_demo['age'].apply(In_age)
```
